```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt = pd.read_csv("alzheimer.csv")
```

```python
dt.head()
```

```python
dt.info()
```

```python
dt.describe()
```

```python
# Eksik değerleri sütun bazında ortalama değer ile doldur
dt['SES'].fillna(dt['SES'].mean(), inplace=True)
dt['MMSE'].fillna(dt['MMSE'].mean(), inplace=True)
```

```python
plt.figure(figsize=(4,4))
sns.countplot(data=dt , x='Group')
plt.title('Hastalığın Evrelerine Göre Dağılımı')
plt.xlabel('Hastalık Durumu')
plt.ylabel('Sayı')
```

```python
plt.figure(figsize=(10,10))
sns.countplot(data=dt , x='Age')
plt.title('Hastalığın Yaşa  Göre Dağılımı')
plt.xlabel('Yaş Durumu')
plt.ylabel('Sayı')
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=dt , x='M/F')
plt.title('Hastalığın Cinsiyetine Göre Dağılımı')
plt.xlabel('Cinsiyet Durumu')
plt.ylabel('Sayı')
```

```python
plt.figure(figsize=(8,8))
sns.countplot(data=dt , x='SES')
plt.title('Hastalığın Sosyoekonomik Duruma  Göre Dağılımı')
plt.xlabel('Hastalık Durumu')
plt.ylabel('Sayı')
```
